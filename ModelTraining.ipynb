```python
 #Feature Selection
DfFinal = DDoS[['Flow Pkts/s','Fwd Pkts/s','Bwd Pkts/s','Init Fwd Win Byts','Label']]
X = DDoS[['Flow Pkts/s','Fwd Pkts/s','Bwd Pkts/s','Init Fwd Win Byts']]
```

```python
sea.heatmap(DfFinal.corr(), annot=True)
```

```python
y = DDoS['Label']
```

```python
y
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
standscale = StandardScaler()
X = standscale.fit_transform(X)
print(X)
```

```python
#Splitting dataset for training and testing
from sklearn.model_selection import train_test_split
X_t, X_ts, y_t, y_ts = train_test_split(X,y, train_size=0.70,test_size=0.30, random_state=11)
```

```python
print(X_t.shape)
print(y_t.shape)
print(X_ts.shape)
print(y_ts.shape)
```

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
modelRFC = RandomForestClassifier(criterion='gini', max_depth=3, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=1, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None)
modelRFC.fit(X_t,y_t)
```

```python
predictionsRFC = modelRFC.predict(X_ts)
```

```python
predictionsRFC
```

```python
y_ts
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, f1_score, recall_score, precision_score

cmatrix = confusion_matrix(y_ts, predictionsRFC)

accuracyRF = accuracy_score(y_ts, predictionsRFC)*100
precisionRF = precision_score(y_ts, predictionsRFC)*100
recallRF = recall_score(y_ts, predictionsRFC)*100
f1RF = f1_score(y_ts, predictionsRFC)*100

print("Accuracy Score:")
print(accuracyRF)
print("------------------------------------------")
print("Precision Score:")
print(precisionRF)
print("------------------------------------------")
print("Recall Score:")
print(recallRF)
print("------------------------------------------")
print("F1 Score:")
print(f1RF)
print("------------------------------------------")
print('\n')
print("Classification Report:")
print(classification_report(y_ts,predictionsRFC))
print('\n')

fg, px = diag.subplots(figsize=(5,5))
px.matshow(cmatrix, cmap=diag.cm.Oranges, alpha=0.3)
for a in range(cmatrix.shape[0]):
    for b in range(cmatrix.shape[1]):
        px.text(x=b, y=a, s=cmatrix[a, b], va='center', ha='center', size='xx-large')

diag.xlabel('Predictions', fontsize=16)
diag.ylabel('Actuals', fontsize=16)
diag.title('Confusion Matrix', fontsize=16)
diag.show()
print("------------------------------------------")
print('\n')



fg = diag.figure(figsize = (10, 7))
px = diag.axes(projection ="3d")

px.scatter3D(y_ts, predictionsRFC, color = "red")
diag.title("RandomForest Classifier")
diag.show()
```
